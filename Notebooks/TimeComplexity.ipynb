# Time test over KendallTau and Spearman correlation functions

In this notebook we will be studying execution time from Kendalltau and Spearman correlation functions from  scipy package

```python
import numpy as np
from scipy.stats import kendalltau, spearmanr
import time as t
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
size = 1000
vector1 = np.random.rand(size)
vector2 = np.random.rand(size)
```

```python
start = t.time()
kendalltau(vector1, vector2)
end = t.time() - start
print(end)
```

```python
sizes = []
times = []

for i in tqdm(range(1, 1000001, 10000)):
    size = i
    vector1 = np.random.rand(size)
    vector2 = np.random.rand(size)
    start = t.time()
    kendalltau(vector1, vector2)
    end = t.time() - start
    sizes.append(size)
    times.append(end)
```

```python
plt.plot(sizes, times)
plt.show()
```

```python
sizes = []
times_spearman = []

for i in tqdm(range(1, 1000001, 10000)):
    size = i
    vector1 = np.random.rand(size)
    vector2 = np.random.rand(size)
    start = t.time()
    spearmanr(vector1, vector2)
    end = t.time() - start
    sizes.append(size)
    times_spearman.append(end)
```

```python
plt.plot(sizes, times_spearman)
plt.show()
```

# Conclusión, la complejidad de ambos métodos parece bastante lineal y si tenemos en cuenta que aumentará respecto al número de filas => La complejidad luce cuadrática
